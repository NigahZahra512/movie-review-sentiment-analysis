```python
import pandas as pd
import re
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
df=pd.read_csv(r"D:\Corvit\Assignments\Sentiment Analysis\02) IMDB Dataset.csv")
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
stop_words=set(stopwords.words("english"))
ps=PorterStemmer()
def clean_text(text):
   text=text.lower()
   text=re.sub(r"<.*?>","",text) #Substitution:HTML 
   text=re.sub(r"http\S+","",text) #URL
   text=re.sub(r"[^a-z0-9\s]","",text) #Punct
   tokens=word_tokenize(text)  #Stemming/StopWords
   cleaned=[ps.stem(w) for w in tokens if w not in stop_words]
   return " ".join(cleaned)
```

```python
df["review"]=df["review"].apply(clean_text)
```

```python
df
```

```python
le=LabelEncoder()
df["sentiment"]=le.fit_transform(df["sentiment"])
```

```python
df
```

```python
49582*5000
```

```python
tfidf=TfidfVectorizer(max_features = 5000)
X=tfidf.fit_transform(df["review"]).toarray()
Y=df["sentiment"].values
```

```python
X,Y
```

```python
X_train,X_test,Y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(f"Data split into {len(X_train)} training rows and {len(X_test)} testing rows.")
```

```python
import torch #used for special arrays
import torch.nn as nn
import torch.optim as optim #Adam Optimizer
from torch.utils.data import TensorDataset, DataLoader
```

```python
Y_train
```

```python
train_ds=TensorDataset(torch.tensor(X_train).float(), torch.tensor(Y_train).float())
```

```python
test_ds=TensorDataset(torch.tensor(X_test).float(), torch.tensor(y_test).float())
```

```python
train_loader=DataLoader(train_ds,batch_size=64,shuffle=True)
```

```python
test_loader=DataLoader(test_ds,batch_size=64)
```

```python
class SentimentRNN(nn.Module):
    def __init__(self,input_size,hidden_size=128):
        super(SentimentRNN,self).__init__()
        self.rnn=nn.RNN(input_size,hidden_size,batch_first=True)
        self.fc=nn.Linear(hidden_size,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self,x):
        _, h_n=self.rnn(x)
        out=self.fc(h_n.squeeze(0))
        return self.sigmoid(out)
```

```python
model=SentimentRNN(input_size=5000)
```

```python
criterion=nn.BCELoss()
```

```python
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
epochs=10
for epoch in range(epochs):
    model.train()
    total_loss=0

    for xb, yb in train_loader:
        xb=xb.unsqueeze(1)
        preds=model(xb).squeeze()
        loss=criterion(preds,yb)
        optimizer.zero_grad()   #clear old math:calculation
        loss.backward()         #calculate update
        optimizer.step()       #update weights
        total_loss+=loss.item()
    print(f"Epoch {epoch+1}/{epochs} | Average Error: {total_loss/len(train_loader): .4f}")
```

```python
epochs=15
for epoch in range(epochs):
    model.train()
    total_loss=0

    for xb, yb in train_loader:
        xb=xb.unsqueeze(1)
        preds=model(xb).squeeze()
        loss=criterion(preds,yb)
        optimizer.zero_grad()   #clear old math:calculation
        loss.backward()         #calculate update
        optimizer.step()       #update weights
        total_loss+=loss.item()
    print(f"Epoch {epoch+1}/{epochs} | Average Error: {total_loss/len(train_loader): .4f}")
```

```python
model.eval()
correct=0
with torch.no_grad():
     for xb, yb in test_loader:
        xb=xb.unsqueeze(1)
        preds=(model(xb).squeeze() > 0.5).float()
        correct+=(preds==yb).sum().item()
print(f"Final Accuracy: {(correct/len(y_test))*100: .2f}%")
```

```python
torch.save(model.state_dict(), "rnn_model.pth")
```
